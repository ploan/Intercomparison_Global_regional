### for insitu dataset

```python
# Intialize empty array to store model timeseries and slope through Thiel Sen

insitu_ann_ts = []
insitu_slope_all = []

# Loop through RCM simulations; store in a dictionary
for i, row in insitu_paths_subset.iterrows():
    
    # Get weighted spatial annual average for time series
    insitu_map = get_map(row[1], variable, time_slice,lat_slice, lon_slice, season, iscale, land_mask, ipcc_mask, ipcc_subregions[region_name])

    insitu_spatial_mask=[]
    for i in range(len(insitu_map.time)): 
        region_xr=insitu_map.isel(time=i)
        a=region_xr.count()
        if ((a/b)>=0.8): 
            region_xr_masked=region_xr
        else: 
            region_xr_masked=region_xr.where(region_xr.values == np.nan)
        insitu_spatial_mask.append(region_xr_masked)
    
    insitu_masked=[*insitu_spatial_mask]
    insitu_map_masked=xr.concat(insitu_masked,dim=('time'))
    trend_time_slice=slice('1985','2016')

    ##calculated weighted data
    weights = np.cos(np.deg2rad(insitu_map_masked.lat))
    weights.name = "weights"
    
    insitu_weighted=insitu_map_masked.weighted(weights)
    insitu_ts=insitu_weighted.mean(dim=('lon','lat'))
    # Calculate trend related to Mann Kendall Metrics based on commom time period
    #insitu_mk_ts = mk.original_test(insitu_ts.sel(time=trend_time_slice))
    
    # Get slope value
    #insitu_slope = insitu_mk_ts[7]
    
    # append time series and slope into array
    insitu_ann_ts.append(insitu_ts)
    #insitu_slope_all.append(insitu_slope)
```

### for satellite dataset


```python
satellite_ann_ts = []
satellite_slope_all = []

# Loop through RCM simulations; store in a dictionary
for i, row in satellite_paths_subset.iterrows():
    
    # Get weighted spatial annual average for time series
    satellite_map_masked = get_map(row[1], variable, time_slice,lat_slice, lon_slice, season, iscale, land_mask, ipcc_mask, ipcc_subregions[region_name])
    satellite_map=satellite_map_masked.where(satellite_map_masked<=200)

    satellite_spatial_mask=[]
    for i in range(len(satellite_map.time)): 
        region_xr=satellite_map.isel(time=i)
        a=region_xr.count()
        if ((a/b)>=0.8): 
            region_xr_masked=region_xr
        else: 
            region_xr_masked=region_xr.where(region_xr.values == np.nan)
        satellite_spatial_mask.append(region_xr_masked)
    
    satellite_masked=[*satellite_spatial_mask]
    satellite_map_masked=xr.concat(satellite_masked,dim=('time'))

    trend_time_slice=slice('1985','2016')

    ##calculated weighted data
    weights = np.cos(np.deg2rad(satellite_map_masked.lat))
    weights.name = "weights"
    satellite_weighted=satellite_map_masked.weighted(weights)

    satellite_ts=satellite_weighted.mean(dim=('lon','lat'))
    
    # Calculate Mann Kendall Metrics based on time series
    #satellite_mk_ts = mk.original_test(satellite_ts.sel(time=trend_time_slice))
    
    # Get slope value
    #satellite_slope = satellite_mk_ts[7]
    
   
    satellite_ann_ts.append(satellite_ts)
    #satellite_slope_all.append(satellite_slope)
```

## for reanalysis data

```python
reanalysis_ann_ts = []
reanalysis_slope_all = []

for i, row in reanalysis_paths_subset.iterrows():
    
    # Get weighted spatial annual average for time series
    reanalysis_map = get_map(row[1], variable, time_slice,lat_slice, lon_slice, season, iscale, land_mask, ipcc_mask, ipcc_subregions[region_name])
    
    reanalysis_spatial_mask=[]
    for i in range(len(reanalysis_map.time)): 
        region_xr=reanalysis_map.isel(time=i)
        a=region_xr.count()
        if ((a/b)>=0.8): 
            region_xr_masked=region_xr
        else: 
            region_xr_masked=region_xr.where(region_xr[variable] == np.nan)
        reanalysis_spatial_mask.append(region_xr_masked)
    
    reanalysis_masked=[*reanalysis_spatial_mask]
    reanalysis_map_masked=xr.concat(reanalysis_masked,dim=('time'))

    trend_time_slice=slice('1985','2016')

    ##calculated weighted data
    weights = np.cos(np.deg2rad(reanalysis_map_masked.lat))
    weights.name = "weights"
    reanalysis_weighted=reanalysis_map_masked.weighted(weights)

    reanalysis_ts=reanalysis_weighted.mean(dim=('lon','lat'))
    
    # Calculate Mann Kendall Metrics based on time series
    #reanalysis_mk_ts = mk.original_test(reanalysis_ts.sel(time=trend_time_slice))
    
    # Get slope value
    #reanalysis_slope = reanalysis_mk_ts[7]
    
    reanalysis_ann_ts.append(reanalysis_ts)
    #reanalysis_slope_all.append(reanalysis_slope)
```

## Calculate relative trend (%/C degree) during 2000-2019

```python
##time series of temperature anomalies 
import xarray as xr
import numpy as np
ds1=xr.open_dataset('/g/data/w40/pn3424/2024/Lisa/Best_monthly_tavg_land_ocean_sett.nc')
##climatology 
clim_month = ds1.tavg.sel(time=slice('2001','2015')).groupby('time.month').mean()
##yearly weighted anomaly 
ano_month = ds1.tavg.groupby('time.month') - clim_month

ano_year=ano_month.groupby('time.year').mean(dim='time')


weights = np.cos(np.deg2rad(ano_year.latitude))
weights.name = "weights"
ano_year_weighted = ano_year.weighted(weights).mean(dim=('latitude','longitude'))
ano_year_weighted
gmst_xr=ano_year_weighted
```

```python
import statsmodels.api as sm
import pandas as pd

model_c=[]
for i in range(len(time_series)):
    if i in range(len(time_series)-1): 
        if (np.isnan(time_series[i].sel(time=slice('2001', '2015')).mean(dim='time').values)==True):
            slope_c = np.nan
        else: 
            pre_clim=time_series[i].sel(time=slice('1985', '2015')).mean(dim='time')
            pre_clim_ano=time_series_ano[i].sel(time=slice('1985','2015'))/pre_clim.values
            
            non_nan_indices = ~np.isnan(pre_clim_ano)
            pr_indx_grid_filtered = pre_clim_ano[non_nan_indices]
            gmst_xr_sel=gmst_xr.sel(year=slice('1985','2015'))

            gmst_xr_masked=[]

            for j in range(len(pre_clim_ano.time)): 
                if (np.isnan(pre_clim_ano.isel(time=j).values)==True):
                    gmst=np.nan
                else: 
                    gmst=gmst_xr_sel.isel(year=j).values
        
                gmst_xr_masked.append(gmst)
            a=np.array(gmst_xr_masked)

        non_nan_indices_temp = ~np.isnan(a)
    
        temp_indx_grid_filtered = a[non_nan_indices_temp]
    
        gmst_lowpass_filtered = sm.add_constant(temp_indx_grid_filtered)

    
        model_c_slope = sm.RLM(pr_indx_grid_filtered.values*100, gmst_lowpass_filtered, M=sm.robust.norms.AndrewWave())
        results_c = model_c_slope.fit()
        slope_c = results_c.params[1]
        model_c.append(slope_c)
    ##regional dataset APHRODITE calculate over 1996-2015
    if (i==(len(time_series)-1)): 
        pre_clim=time_series[i].sel(time=slice('2001', '2015')).mean(dim='time')
        ##
        pre_clim_ano=time_series_ano[i].sel(time=slice('1985','2015'))/pre_clim.values
            
        non_nan_indices = ~np.isnan(pre_clim_ano)
        pr_indx_grid_filtered = pre_clim_ano[non_nan_indices]
        gmst_xr_sel=gmst_xr.sel(year=slice('1985','2015'))

        gmst_xr_masked=[]

        for j in range(len(pre_clim_ano.time)): 
            if (np.isnan(pre_clim_ano.isel(time=j).values)==True):
                gmst=np.nan
            else: 
                gmst=gmst_xr_sel.isel(year=j).values
        
            gmst_xr_masked.append(gmst)
            

        a=np.array(gmst_xr_masked)

        non_nan_indices_temp = ~np.isnan(a)
    
        temp_indx_grid_filtered = a[non_nan_indices_temp]
    
        gmst_lowpass_filtered = sm.add_constant(temp_indx_grid_filtered)

    
        model_c_slope = sm.RLM(pr_indx_grid_filtered.values*100, gmst_lowpass_filtered, M=sm.robust.norms.AndrewWave())
        results_c = model_c_slope.fit()
        slope_c = results_c.params[1]
        model_c.append(slope_c)
```

```python
relative_trend=np.array(model_c) 
relative_trend
```
